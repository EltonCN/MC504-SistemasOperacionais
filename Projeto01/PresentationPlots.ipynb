```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.sin(np.linspace(0,10,100))
noise = np.random.normal(scale=0.5,size=100)

fig = plt.figure()
plt.title("Signal")
plt.xlabel("Time")
plt.plot(y)
plt.show()
fig.savefig("Signal.png", facecolor="white", transparent=False)

fig = plt.figure()
plt.title("Signal+Noise")
plt.xlabel("Time")
plt.plot(y+noise)
plt.show()
fig.savefig("Noise.png", facecolor="white", transparent=False)
```

```python
import scipy.stats as stats

fig = plt.figure()
x = np.linspace(-3, 3, 100)
plt.plot(x, stats.norm.pdf(x, 0, 1))
fig.savefig("Normal.png", facecolor="white", transparent=False)
```
